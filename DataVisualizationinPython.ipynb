# Data Visualization in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_1 = np.array([2,5,4,7,6,1,3,8])
array_2 = np.array([4,2,3,1,4,3,1,2])
```

### Overlaying two arrays

```python
#plt.plot(x, y)
plt.plot(array_1)
plt.plot(array_2)
plt.xlabel('Time')
plt.title('Two Dependent Variables Over Time')
plt.show()
```

### Making horizontal subplots

```python
plt.axes([.05, .05, .425, .9]) # plt.axes(x low, y low, width, height) x low must be .6, .5 squishes graphs on x axis
plt.plot(array_1)
plt.xlabel('Time')
plt.title('Dependent Variable 1')
plt.axes([.6, .05, .425, .9])
plt.plot(array_2)
plt.xlabel('Time')
plt.title('Dependent Variable 2')
plt.show()
```

### Making subplots that are vertically stacked

```python
plt.subplot(2,1,1) #plt.subplot(num rows, num columns, num subplot)
plt.plot(array_1)
plt.xlabel('Time')
plt.title('Dependent Variable 1')
plt.subplot(2,1,2)
plt.plot(array_2)
plt.xlabel('Time')
plt.title('Dependent Variable 2')
plt.tight_layout()
plt.show()
```

### Setting axis limits

```python
#plt.axis([xmin, xmax, ymin, ymax])
#plt.xlim([xmin, xmax])
#plt.ylim([ymin, ymax])
array_3 = np.array([5,3,8,7,9,10,2,5,11,12])
plt.plot(array_3)
plt.xlabel('Time')
plt.title('Dependent Variable 1')
plt.xlim([0,10])
plt.show()
```

```python
array_3 = np.array([5,3,8,7,9,10,2,5,11,12])
plt.plot(array_3)
plt.xlabel('Time')
plt.title('Dependent Variable 1')
plt.xlim([0,6])
plt.show()
```

```python
#plt.axis('off') turns off axis
#plt.axis('equal') scales x and y axis equally
#plt.axis('square') forces output into a square
#plt.axis('tight') makes sure all data is included in graph
plt.plot(array_3)
plt.xlabel('Time')
plt.title('Dependent Variable 1')
plt.xlim([0,6])
plt.axis('square')
plt.show()
```

### Adding a legend

```python
time = np.array([1,2,3,4,5,6,7,8])
array_4 = np.array([2,4,5,3,7,6,9,6])
plt.scatter(time, array_1, marker='o', color='red', label='Array 1')
plt.scatter(time, array_2, marker='*', color='purple', label='Array 2')
plt.scatter(time, array_4, marker='+', color='orange', label='Array 3')
plt.legend(loc='upper left')
plt.show()
```

### Annotate over graph

```python
time = np.array([1,2,3,4,5,6,7,8])
array_4 = np.array([2,4,5,3,7,6,9,6])
plt.scatter(time, array_1, marker='o', color='red')
plt.annotate('Num 1', xy=(3,5))
plt.scatter(time, array_2, marker='*', color='purple')
plt.annotate('Num 2', xy=(6,1))
plt.scatter(time, array_4, marker='+', color='orange')
plt.annotate('Num 3', xy=(7,5))
plt.show()
```

### Add arrows

```python
time = np.array([1,2,3,4,5,6,7,8])
array_4 = np.array([2,4,5,3,7,6,9,6])
plt.scatter(time, array_1, marker='o', color='red')
plt.annotate('Num 1', xy=(3,5), xytext=(2,4), arrowprops={'color':'blue'})
plt.scatter(time, array_2, marker='*', color='purple')
plt.annotate('Num 2', xy=(6,1), xytext=(4,5), arrowprops={'color':'green'})
plt.scatter(time, array_4, marker='+', color='orange')
plt.annotate('Num 3', xy=(7,5), xytext=(5,6), arrowprops={'color':'orange'})
plt.show()
```
