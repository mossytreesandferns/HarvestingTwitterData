## Analysis of Variance (ANOVA) of Median House Value Based on Location

```python
housing['ocean_proximity'].unique()
```

```python
housing.set_index(['ocean_proximity']).head()
```

```python
near_bay = housing.loc[housing['ocean_proximity']=='NEAR BAY']
one_h_ocean = housing.loc[housing['ocean_proximity']=='<1H OCEAN']
inland = housing.loc[housing['ocean_proximity']=='INLAND']
near_ocean = housing.loc[housing['ocean_proximity']=='NEAR OCEAN']
island = housing.loc[housing['ocean_proximity']=='ISLAND']
```

### H0: There is no difference in the means of median house values of different housing locations.
### HA: There is a difference in at least one of the means.

```python
near_bay['median_house_value'].head()
```

```python
from scipy import stats
import statsmodels.api as sm
import researchpy as rp
from statsmodels.formula.api import ols

f, p = stats.f_oneway(near_bay['median_house_value'],one_h_ocean['median_house_value'],inland['median_house_value'],near_ocean['median_house_value'],island['median_house_value'])
print('The f statistic is:', f, ' The p-value is:', p)
```

#### With an assumed alpha on .05, we reject the null hypotheses in favor of the alternative.  The evidence suggests there is 
#### a statistically significant difference between at least one mean of median house value and the rest.
