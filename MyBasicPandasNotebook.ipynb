# Dataframes (matrices or stacks of them)

```python
# Construct a dictionary of series (note 'one and 'two' will become column names)
df = {'one': pd.Series([1,3,5,7], index = ['even','nums','mod','zero']),
           'two': pd.Series([2,4,6,8], index = ['even','numbrs','mods','zero'])}
```

```python
# Create a dataframe from your dictionary
# Notice that indices are alphabetical in a single column while values are separated into their named columns
dframe1 = pd.DataFrame(df)
print(dframe1)
```

```python
# Call the index values
dframe1.index
```

```python
# Call the dataframe columns
dframe1.columns
```

```python
# Call the values of a few indices
pd.DataFrame(df, index = ['mod','mods','nums','zero'])
```

```python
# Call values of a few indices and two columns, including an imaginary one
pd.DataFrame(df, index=['mod','mods','nums','zero'], columns=['one','five'])
```

## Interacting with Timestamps

```python
# Import time and tag data
tags = pd.read_csv('/Users/meganporter/EDXPy4DS/Week-4-Pandas/movielens/ml-20m/tags.csv', sep=',')
tags.head()
```

```python
tags.dtypes
```

```python
# Using a function to convert timestamp to user-readable time
tags['parsed_time'] = pd.to_datetime(tags['timestamp'], unit='s')
tags['parsed_time'].dtype
```

```python
tags.head(5)
```

```python
# Select rows later than February, 1, 2015
greater_than_t = tags['parsed_time'] > '2015-02-01'

selected_rows = tags[greater_than_t]

tags.shape, selected_rows.shape
```

```python
# Sort tag data frame in ascending order by parsed_time
tags.sort_values(by='parsed_time', ascending=True)[:10]
```

```python
average_rating = ratings[['movieId','rating']].groupby('movieId', as_index=False).mean()
average_rating.tail()
```

```python
# Join movies to newly created average_rating
joined = movies.merge(average_rating, on='movieId', how='inner')
joined.head()
joined.corr()
```

```python
yearly_average = joined[['year','rating']].groupby('year', as_index=False).mean()
yearly_average[:10]
```

```python
yearly_average[-20:].plot(x='year', y='rating', figsize=(15,10), grid=True)
```
