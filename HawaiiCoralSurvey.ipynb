### Biomass of the top ten counted species of fish

```python
fish = fish_biomass['TAXONNAME'].value_counts()
fish.head(10)
```

```python
fishes = fish_biomass[['TAXONNAME', 'Biomass']]
fishes.head()
```

```python
fishes.sort_values(by=['Biomass']).head()
```

# Aggregate Function in Pandas

```python
fish1 = fishes.groupby('TAXONNAME')['Biomass'].sum()
fish1.head(10)
```

```python
print(type(fish1))
```

```python
fish2 = fish1.to_frame(name=None).head(10)
fish2.head(10)
```

```python
fish3 = fish2.sort_values(by=['Biomass'],ascending=False)
fish3
```

```python
fish3.columns
```
