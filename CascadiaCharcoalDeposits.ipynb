### I wonder how charcoal count and concentration changes over time, if at all?

```python
time = data['Year AD']
char_count = data['Charcoal (count)']
char_concentration = data['Concentration (#/ml)']
char_volume = data['volume (ml)']
```

```python
plt.figure(figsize=(20,15))
plt.plot(time, char_count, color='b', label = "Charcoal Count")
plt.plot(time, char_concentration, color='orange', label = "Charcoal Concentration")
plt.plot(time, char_volume, color= 'm', label = "Charcoal Volume")
plt.legend()
plt.xlabel('Time')
plt.ylabel('Charcoal Deposits')
plt.title('Charcoal Count, Concentration, and Volume Over Time')
plt.show()
```
